```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.linspace(0, 10, 100)
Y1, Y2 = np.sin(X), np.cos(X)
fig, ax = plt.subplots()
ax.plot(X, Y1, X, Y2)
plt.show()
```

```python
X = np.linspace(0, 10, 100)
Y1, Y2 = np.sin(2*X), np.cos(X)
fig, ax = plt.subplots()
ax.plot(X, Y1, X, Y2)
plt.show()
```

```python
X = np.linspace(0, 10, 100)
Y1, Y2, Y3 = np.sin(2*X), np.cos(X), (np.tan(X))/5
fig, ax = plt.subplots()
ax.plot(X, Y1, X, Y2, X, Y3)
plt.show()
```

```python
data = [0.27, 0.67, 0.06]
labels = 'Low', 'Medium', 'High'
colors = ['yellowgreen', 'gold', 'lightskyblue']
plt.pie(data, explode=(0,0,0), labels=labels, colors=colors, autopct='%1.1f%%', shadow=False)
plt.show()
```
